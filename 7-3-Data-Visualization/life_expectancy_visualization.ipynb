```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 데이터 로드
df = pd.read_csv('./Life Expectancy.csv')
```

```python

# 데이터 살펴보기
df.head()
```

```python
df.info()
# print(df.columns)
```

```python
# 소득 수준과 기대 수명 간의 관계 시각화
plt.figure(figsize=(12, 6))
sns.scatterplot(x='GDP', y='Life expectancy ', data=df, alpha=0.7)
plt.title('Relationship between GDP and Life Expectancy')
plt.xlabel('GDP')
plt.ylabel('Life Expectancy')
plt.show()
```

```python
# 소득 수준에 따른 기대 수명 추세선 추가
plt.figure(figsize=(12, 6))
sns.regplot(x='GDP', y='Life expectancy ', data=df, scatter_kws={'alpha':0.5})
plt.title('Trend of GDP and Life Expectancy')
plt.xlabel('GDP')
plt.ylabel('Life Expectancy')
plt.show()
```

```python
# 연도별 기대 수명 변화 시각화
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Life expectancy ', data=df, errorbar=None)
plt.title('Trend of Life Expectancy over Years')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.show()
```

```python
# Explicit content 여부가 기대 수명에 미치는 영향 시각화 (여기서는 Status 사용)
plt.figure(figsize=(12, 6))
sns.boxplot(x='Status', y='Life expectancy ', data=df)
plt.title('Life Expectancy by Status')
plt.xlabel('Status')
plt.ylabel('Life Expectancy')
plt.show()
```
