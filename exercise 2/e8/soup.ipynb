```python
from urllib.request import urlopen
```

```python
URL = "https://en.wikipedia.org/wiki/HyperCard"

res = urlopen(URL)
body = res.read()
body
```

```python
soup2 = BeautifulSoup(body)
soup2
```

```python
soup2.find_all(name="p")[0].__class__
```

```python
for line in map(lambda x: x.text, soup2.find_all(name="p")):
    if (line := line.strip()):
        print(line)
```
